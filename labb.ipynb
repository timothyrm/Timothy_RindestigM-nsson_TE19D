```python
#Uppgift E

import math
import random
import matplotlib.pyplot as plt

plt.xlabel("x")
plt.ylabel("y")

x = -1
y = 1

avstånd=0
antal=10

for i in range(antal):
    x=random.uniform(-1,1)
    y=random.uniform(-1,1)  
    if avstånd < 1: 
        avstånd += 1
        plt.plot(x,y,'*r')
    else:
        plt.plot(x,y,'*b')
print(f"{avstånd/antal}")

#Jag har tagit samma kod från uppgiften innan men ändrat om och gjort en if sats där man tar roten ur x^2 och y^2
```

```python
#Uppgift H 

import random
import math
import matplotlib.pyplot as plt

punktercirkel=0
antal = 5000

for i in range(antal):
    x=random.uniform(-1,1)
    y=random.uniform(-1,1)
    print(f"{x,y}")
    if math.sqrt(x**2 + y**2) < 1:
        punktercirkel += 1
        plt.plot(x,y,'*r')*2
    else:
        plt.plot(x,y,'*g')

#radiens förändring påverkar inte cirkeln så mycket, det blir bara fler prints och fler punkter. Dock ifall man skriver in fel kod så kommer Y multipliceras med radien och då kommer cirkelns höjd förändras. 
```
